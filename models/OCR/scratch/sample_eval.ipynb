```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image
```

```python
df = pd.read_csv("mnist_test_final.csv")

imgs = df.drop(columns=["label"])
labels = df["label"]
```

```python
images = 9

plt.figure(figsize=(10, 11))

defaults = []
for i in range(images):
    img = imgs.iloc[i].values.reshape(28,28) # all images are 28 * 28 pixels
    defaults.append(Image.fromarray(img.astype(np.uint8)))
    plt.subplot(3, 3, i+1)
    plt.imshow(img, cmap="gray")
    plt.title(str(labels.iloc[i]))
```

```python
transformation = transforms.Compose([
    transforms.Grayscale(3),
    transforms.Resize(256)
])
```
