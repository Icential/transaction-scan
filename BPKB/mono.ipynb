```python
import cv2
import matplotlib.pyplot as plt
```

```python
bpkb = cv2.imread("bpkb_p2.png")
bpkb = cv2.cvtColor(bpkb, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(30,8))
plt.imshow(bpkb, cmap="gray")
```

```python
plt.figure(figsize=(30,8))
plt.imshow(cv2.threshold(bpkb, 150, 255, cv2.THRESH_BINARY)[1], cmap="gray")
```

```python
plt.figure(figsize=(30,8))
plt.imshow(cv2.adaptiveThreshold(bpkb, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 27), cmap="gray")
```
